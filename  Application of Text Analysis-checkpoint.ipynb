```python
# loading the file

myfile = open(r"D:\ML ALOGRITHMS\reddy.txt")
txt = myfile.read()
print(txt)
```

```python
# import libraries
import nltk
from nltk import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk import FreqDist
import numpy as np
import pandas as pd
import re
```

```python
#tokenization
sent = nltk.sent_tokenize(txt)
```

```python
sent
```

```python
sent = nltk.word_tokenize(txt)
sent
```

```python
sent.count
```

```python
#Lemmatization
wordnet = WordNetLemmatizer()
corpus=[]
for i in range(len(sent)):
    review = re.sub('[^a-zA-Z]','',sent[i])

    review= review.lower()
    review=review.split()
    review=[wordnet.lemmatize(word) for word in review if not word in set(stopwords.words('english'))]
    review=' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
#Frequency distribution chart
fdist = FreqDist(corpus)
fdist.plot(10,title='Frequency distribution')
```
